```python
import cv2
import numpy as np
```

```python
# Resmi yükle
image = cv2.imread("denemeci0.jpg")

# 1. Renk Algılama
# Renk aralığını tanımla (mavi renk için)
lower_color = np.array([100, 150, 1])  # Alt renk aralığı (HSV formatında)
upper_color = np.array([140, 255, 255])  # Üst renk aralığı (HSV formatında)

# Görüntüyü HSV formatına çevir
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Belirtilen aralığın içindeki piksell bul
mask = cv2.inRange(hsv_image, lower_color, upper_color)

# Maskedeki piksell sayısını bul
pixel_count = cv2.countNonZero(mask)
print(f"Belirlenen renk aralığında {pixel_count} piksel bulundu.")

# Maskeyi göster
cv2_imshow(mask)

# 2. Kenar Algılama
# Kenar algılama için Canny algoritması kullanma
edges = cv2.Canny(image, 100, 200)

# Kenarları göster
cv2_imshow(edges)

# Pencereleri kapatmak için bir tuşa basmayı bekle
cv2.waitKey(0)
cv2.destroyAllWindows()
```
